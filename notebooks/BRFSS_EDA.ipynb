```python
import matplotlib.pyplot as plt

gender_filtered = df[df['Response'].isin(['Female', 'Male'])]

gender_totals_filtered = gender_filtered.groupby('Response')['Sample_Size'].sum()

gender_totals_filtered

sex_distribution = gender_totals_filtered

# Plot the data
plt.figure(figsize=(10, 6))
sex_distribution.plot(kind='bar', color=['pink', 'blue'])
plt.title('Sex Distribution', fontsize=16, weight='bold')
plt.xlabel('Sex')
plt.ylabel('Sample Size')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('../figures/brfss_chloe/sex_distribution.png', format='png', bbox_inches='tight')
plt.show()
```

```python
state_totals = df.groupby('Locationdesc')['Sample_Size'].sum()

plt.figure(figsize=(14, 8))
state_totals.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Total Sample Size by State', fontsize=16, weight='bold')
plt.xlabel('State')
plt.ylabel('Sample Size')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('../figures/brfss_chloe/sampleSize_State.png', format='png', bbox_inches='tight')
plt.show()
```

```python
age_totals = df[df['Break_Out_Category'] == 'Age Group'].groupby('Break_Out')['Sample_Size'].sum()

plt.figure(figsize=(10, 6))
age_totals.sort_index().plot(kind='bar', color='orange')
plt.title('Total Sample Size by Age Group', fontsize=16, weight='bold')
plt.xlabel('Age Group')
plt.ylabel('Sample Size')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('../figures/brfss_chloe/sampleSize_Age.png', format='png', bbox_inches='tight')
plt.show()
```

```python
import seaborn as sns
race_count_data = df[df['Break_Out_Category'] == 'Race/Ethnicity'][['Break_Out', 'Sample_Size']]

race_count_data = race_count_data.dropna()

grouped_race_count = race_count_data.groupby('Break_Out')['Sample_Size'].sum().reset_index()

plt.figure(figsize=(12, 8))
sns.barplot(
    data=grouped_race_count,
    x='Sample_Size',
    y='Break_Out',
    palette='coolwarm'
)

plt.title('Population Count by Race/Ethnicity', fontsize=16, weight='bold')
plt.xlabel('Total Count (Sample Size)', fontsize=14)
plt.ylabel('Race/Ethnicity', fontsize=14)
plt.tight_layout()
plt.savefig('../figures/brfss_chloe/popCount_Race_Ethnicity.png', format='png', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Break_Out_Category', palette='Set2')
plt.title('Distribution of Breakout Categories',fontsize=16,weight='bold')
plt.xlabel('Breakout Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.savefig('../figures/brfss_chloe/breakoutCategories.png', format='png', bbox_inches='tight')
plt.show()
```

```python
grouped_job_data = df.groupby(['Break_Out', 'Locationdesc'], as_index=False)['Sample_Size'].sum()

# Pivot the data for plotting (each Break_Out category becomes a row, and we plot each 'Sample_Size' as a separate segment)
pivot_job_data = grouped_job_data.pivot_table(index='Break_Out', columns='Locationdesc', values='Sample_Size', aggfunc='sum').fillna(0)

# Plot the stacked bar plot
fig, ax = plt.subplots(figsize=(10, 7))
pivot_job_data.plot(kind='barh', stacked=True, ax=ax, colormap='tab20', edgecolor='black')

# Set titles and labels
ax.set_title('Household Income Distribution', fontsize=16, weight='bold')
ax.set_xlabel('Total Population (Sample Size)', fontsize=14)
ax.set_ylabel('Household Income Category', fontsize=14)
ax.legend().set_visible(False)
# Adjust layout
plt.tight_layout()
plt.savefig('../figures/brfss_chloe/HouseholdIncome.png', format='png', bbox_inches='tight')
plt.show()
```

```python
smoking_data = df[(df['Break_Out_Category'] == 'Race/Ethnicity') & 
                    (df['Response'].str.contains('smoke', case=False, na=False))][['Break_Out', 'Sample_Size', 'Response']]

print(smoking_data.head())

smoking_data = smoking_data.dropna()
grouped_smoking_data = smoking_data.groupby(['Break_Out', 'Response'])['Sample_Size'].sum().reset_index()
pivot_smoking_data = grouped_smoking_data.pivot(index='Break_Out', columns='Response', values='Sample_Size').fillna(0)
pivot_smoking_data.plot(kind='barh', stacked=True, figsize=(12, 8), colormap='viridis', edgecolor='black')

plt.title('Smoking Responses by Race/Ethnicity', fontsize=16, weight='bold')
plt.xlabel('Sample Size (Count)', fontsize=14)
plt.ylabel('Race/Ethnicity', fontsize=14)
plt.legend(title='Smoking Responses', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
plt.tight_layout()
plt.savefig('../figures/brfss_chloe/SmokingResponses_RaceEthnicity.png', format='png', bbox_inches='tight')
plt.show()
```

```python
from geopy.geocoders import Nominatim
import time
```

```python
geolocator = Nominatim(user_agent="city_locator")
```

```python
def get_city(lat, lon):
    try:
        location = geolocator.reverse((lat, lon), language='en', timeout=10)
        address = location.raw.get('address', {})
        city = address.get('city', None) or address.get('town', None) or address.get('village', None)
        return city
    except Exception as e:
        print(f"Error with lat: {lat}, lon: {lon} -> {e}")
        return None

# Apply function to the DataFrame
merged_df['city'] = encoded_df.apply(lambda row: get_city(row['latitude'], row['longitude']), axis=1)
```

```python
encoded_df.to_csv('BRFSS_GeoLocation.csv', index=False)
```

### The below creates min, max and mean columns - but assumes we are carring about facility

```python
columns_to_calc = ['Sample_Size', 'Confidence_limit_Low', 'Confidence_limit_High']

for col in columns_to_calc:
    merged_df[f'{col}_min'] = merged_df[col].min()
    merged_df[f'{col}_max'] = merged_df[col].max()
    merged_df[f'{col}_mean'] = merged_df[col].mean()
cols = columns_to_calc + ['Year', 'Display_order']
merged_df = merged_df.drop(columns=cols)
merged_df
```

```python
df_cleaned = merged_df.dropna(subset=['GeoLocation'], axis=0).drop_duplicates()
df_cleaned
```

```python
df_cleaned.to_csv('BRFSS_GeoLocation.csv', index=False)
```
