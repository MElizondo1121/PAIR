```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
urban_rural_label1 = pd.read_csv('../CENSUS/2020_Census_ua_st_list_all.csv')
urban_rural_label1['PART'] = urban_rural_label1['PART'].fillna(0)
urban_rural_label1
```

```python
urban_rural_label1['NAME'] = urban_rural_label1['NAME'].str.split(',').str[0]
urban_rural_label1['NAME'] = urban_rural_label1['NAME'].str.replace(r',.*', '', regex=True)
```

```python
print(urban_rural_label1.shape)
urban_rural_label1['PART'].value_counts()
```

```python
texas_data1 = urban_rural_label1[urban_rural_label1['STATE'] == 'Texas']
texas_data1
```

```python
print(texas_data1.shape)
texas_data1['PART'].value_counts()
```

```python
texas_data1.to_csv('2020_Census_ua_st_list_all_TEXAS.csv')
```

```python
texas_data1.columns
```

```python
cols = ['STATE', 'UACE', 'STATEFP', 'PART']
df_copy = texas_data1.drop(cols, axis=1)
```

```python
df_copy
```

```python
columns_to_clean  = ['HOUSING', 'AREALAND', 'AREAWATER', 'POPDEN']

df_copy.loc[:, 'HOUSING'] = df_copy['HOUSING'].str.replace(',', '').astype(float)
df_copy.loc[:, 'AREALANDSQMI'] = df_copy['AREALANDSQMI'].str.replace(',', '').astype(float)
df_copy.loc[:, 'AREAWATER'] = df_copy['AREAWATER'].str.replace(',', '').astype(float)
df_copy.loc[:, 'POPDEN'] = df_copy['POPDEN'].str.replace(',', '').astype(float)
```

```python
df_copy.columns
```

```python
plot = ['POP', 'HOUSING', 'AREALAND', 'AREALANDSQMI', 'AREAWATER',
       'AREAWATERSQMI', 'POPDEN']
```

```python
sns.pairplot(df_copy[plot])
plt.suptitle('Pairplot of Census Urban Data', y=1.02, fontsize=16)
plt.tight_layout()
plt.show()
```

```python
corr = df_copy[plot].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation Matrix of CENSUS Urban data', fontsize=16)
plt.tight_layout()
plt.show()
```
